# UNIVARIATE ANALYSIS

## DISTRIBUTION ANALYSIS

```python
plt.figure(figsize=(18, 10))

plt.subplot(2, 3, 1)
sns.histplot(df["discounted_price"], kde=True, color='skyblue')
plt.title("Discounted Price Distribution")

plt.subplot(2, 3, 2)
sns.histplot(df["actual_price"], kde=True, color='lightgreen')
plt.title("Actual Price Distribution")

plt.subplot(2, 3, 3)
sns.histplot(df["discount_percentage"], kde=True, color='salmon')
plt.title("Discount Percentage Distribution")

plt.subplot(2, 3, 4)
sns.histplot(df["rating"], bins=20, color='orange')
plt.title("Rating Distribution")

plt.subplot(2, 3, 5)
sns.histplot(df["rating_count"], bins=30, color='purple')
plt.yscale("log")
plt.title("Rating Count Distribution (Log Scale)")

plt.tight_layout()
plt.show()
```

# CATEGORY-WISE ANALYSIS

## top 10 categories

```python
top_categories = df['category'].value_counts().nlargest(10).index
df_top = df[df['category'].isin(top_categories)]
```

### Count of products in top categories

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_top, y='category', order=top_categories, palette='viridis')
plt.title("Top 10 Product Categories by Count")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

## Category-wise average metrics

```python
category_stats = df_top.groupby("category")[["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]].mean().sort_values("discounted_price", ascending=False)
```

## Avg discounted price

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=category_stats.index, y=category_stats["discounted_price"], palette="coolwarm")
plt.title("Average Discounted Price by Category")
plt.ylabel("Avg Price (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Avg rating per category

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=category_stats.index, y=category_stats["rating"], palette="mako")
plt.title("Average Rating by Category")
plt.ylim(3.5, 5)
plt.ylabel("Avg Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# BIVARIATE RELATIONSHIPS

## Discount % vs Rating

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_top, x="discount_percentage", y="rating", hue="category", alpha=0.7, legend=False)
plt.title("Discount Percentage vs Rating")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Price vs Rating Count (log scale)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=np.log1p(df_top["actual_price"]),
    y=np.log1p(df_top["rating_count"]),
    hue=df_top["category"],
    alpha=0.6,
    legend=False
)
plt.title("Log(Actual Price) vs Log(Rating Count)")
plt.xlabel("Log(Actual Price)")
plt.ylabel("Log(Rating Count)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# CORRELATION HEATMAP

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df[["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]].corr(),
    annot=True, cmap="coolwarm", fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
